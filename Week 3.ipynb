```python
#Importing Libraries

import requests 
import pandas as pd 
import numpy as np 
import random
from geopy.geocoders import Nominatim 

# libraries for displaying images
from IPython.display import Image 
from IPython.core.display import HTML 
from IPython.display import display_html

    
# tranforming json file into a pandas dataframe library
from pandas.io.json import json_normalize


import folium
from bs4 import BeautifulSoup as soup
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors

print('Folium installed')
print('Libraries imported.')
```

### Raw data

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
raw_data=soup(url,'lxml')

table_1 = str(raw_data.table)
display_html(table_1,raw=True)
```

```python
df = pd.read_html(table_1)
df = df[0]
df.head()
```

```python
df1 = df[df.Borough != 'Not assigned']
df2 = df1.groupby(['Postal Code'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

df2.shape
```

```python
geo_data = pd.read_csv('https://cocl.us/Geospatial_data')
geo_data
```

```python
df3 = pd.merge(df2,geo_data,on='Postal Code')
df3.head()
```

```python
latitude = df3['Latitude']
longitude = df3['Longitude']

# create map of Toronto using latitude and longitude values
Toronto_map = folium.Map(location= [43.6532, -79.3832], zoom_start=5)

# add markers to map
for lat, lng, label in zip(latitude, longitude, df3['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Toronto_map)
    
Toronto_map
```

```python
k=5
toronto_clustering = df3.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df3.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df3
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=5)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(latitude, longitude, df3['Neighbourhood'], df3['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
